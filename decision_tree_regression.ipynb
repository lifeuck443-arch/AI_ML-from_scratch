```python
# 1. Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
# 2. Sample data (feature: X, target: y)
X = np.array([
        [1],
        [2],
        [3],
        [4],
        [5]
    ])
y = np.array([1.2, 1.9, 3.2, 3.9, 5.1])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
# 3. Create and train model
model = DecisionTreeRegressor(max_depth=3, random_state=0)
model.fit(X, y)

# 4. Make predictions
X_test 
y_pred = model.predict(X_test)

# 5. Visualize results
plt.scatter(X, y, color="red", label="Actual Data")
plt.plot(X_test, y_pred, color="blue", label="Decision Tree Prediction")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Decision Tree Regression Example")
plt.legend()
plt.show()
```

```python
from sklearn import tree
plt.figure(figsize=(10,6))
tree.plot_tree(model, filled=True, feature_names=["X"], rounded=True)
plt.show()
```
